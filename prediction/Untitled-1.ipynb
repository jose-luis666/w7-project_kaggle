```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
# Concatenar train y test
df = pd.concat([train, test], ignore_index=True)

# Separar en X e Y
Y = df['Price']
X = df.drop('Price', axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# Elimino Operating System Version por los nulos

df = df.drop("Operating System Version", axis=1)
```

```python
df.info()
```

```python
df.rename(columns={' Storage': 'Storage'}, inplace=True)
```

```python
df['Screen Size'] = df['Screen Size'].str.replace('"', '')
df['Screen Size'].value_counts()
```

```python
df['RAM'] = df['RAM'].str.replace('GB', '')
df['RAM'].value_counts()
```

```python
df['Model Name'].value_counts()
```

```python
# Elimino Model Name porque al hacer value counts, vemos que son valores muy diferentes, y no nos sirven de mucho para nuestro estudio.

df = df.drop("Model Name", axis=1)
```

```python
df['Touchscreen'] = 0

for i, valor in enumerate(df['Screen']):
    if 'Touchscreen' in valor:
        df.at[i, 'Touchscreen'] = 1
```

```python
df['CPU'] = df['CPU'].str.split().str[0]

df['CPU'].value_counts()
```

```python
df.head()
```

```python
conversiones = {
    3 : ['2880x1800', '3840x2160', '3200x1800', '3200x1800'],
    2 : ['2400x1600', '2560x1600', '2560x1440', '2736x1824'],
    1 : ['1920x1200', '1920x1080', '2256x1504', '2304x1440', '2160x1440'],
    0 : ['1366x768', '1440x900', '1600x900']
}

# Iterar sobre los valores de la columna 'columna_objeto'
for i, valor in enumerate(df['Screen']):
    for conversion, resoluciones in conversiones.items():
        if any(resolucion in valor for resolucion in resoluciones):
            df.at[i, 'Screen'] = conversion
            break


df['Screen'].value_counts()
```

```python
import pandas as pd

# Limpiar la columna 'Storage'
df['Storage'] = df['Storage'].str.replace(r'\D', '', regex=True)
df.rename(columns={'Storage': 'Storage GB'}, inplace=True)
df.head()
```

```python


# Limpiar la columna 'GPU'
df['GPU'] = df['GPU'].str.replace(r'\d', '', regex=True)

# Mostrar el resultado
df.head()
```

```python
# Limpiar la columna 'Weight'
df['Weight'] = df['Weight'].str.replace(r'\D', '', regex=True)

df.head()
```

```python
df.rename(columns={'Weight': 'Weight kg'}, inplace=True)
```

```python
df.head()
```

```python
df = df.drop("Weight kg", axis=1)
df.head()
```

```python
df.dtypes
```

```python
df['Screen Size'] = df['Screen Size'].astype(float)
df['Screen'] = df['Screen'].astype(int)
df['RAM'] = df['RAM'].astype(int)
df['Storage GB'] = df['Storage GB'].astype(int)
```

```python
df.dtypes
```

```python
columns_to_encode = ['Manufacturer', 'Category', 'CPU', 'GPU', 'Operating System']

encoded_data = pd.get_dummies(df, columns = columns_to_encode)
```
